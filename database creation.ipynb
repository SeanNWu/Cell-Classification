Code Below is the function to iterate through all my CIFs file. Needs change based on Dr. Billinge new Schema

```python
from __future__ import print_function
from diffpy.structure import loadStructure
from pyobjcryst import loadCrystal
from diffpy.srreal.bondcalculator import BondCalculator
from diffpy.structure.parsers import p_cif
from pyobjcryst import loadCrystal

def getlists(filepath):
    a =  p_cif.P_cif()
    a.parseFile(filepath)
    nacl_stru = loadStructure(filepath)
    nacl_objc = loadCrystal(filepath)
    
    bc = BondCalculator()
    
    return nacl_objc,bc(nacl_stru), bc(nacl_objc), nacl_objc.GetSpaceGroup()
```

The Cell Below reads a CIF diffraction pattern and computes the distances list between each atomic pair

```python
from diffpy.structure import loadStructure
from diffpy.srreal.bondcalculator import BondCalculator

structure = loadStructure('/Users/seanwu/Desktop/Full Set of Experimental Patterns (785 CIFs)/wm6137Isup2.rtv.combined.cif')

bond_calculator = BondCalculator()
max_distance = 5  
bond_calculator.rmax = max_distance
bond_calculator.eval(structure)

labels = [atom.element + str(i) for i, atom in enumerate(structure)]

atom_pairs = bond_calculator.sites0, bond_calculator.sites1
distances = bond_calculator.distances

for atom1_idx, atom2_idx, d in zip(*atom_pairs, distances):
    atom1 = labels[atom1_idx]
    atom2 = labels[atom2_idx]
    print(f"{atom1}-{atom2}: {d:.3f} Å")
```

Code Below is Looping through the directory of CIFs file and writing a Json File

```python
import os
import random
import tqdm

directory = '/Users/seanwu/Desktop/Full Set of Experimental Patterns (785 CIFs)'
file_list = os.listdir(directory)
random.shuffle(file_list)
IDLIST = []
ExpandedUnitCellBondDistances = []
CrystalCellBondDistances = []
SpaceGroup = []
count = 0

with tqdm.tqdm(total=len(file_list)) as pbar:
    for filename in file_list:
        if count < 20:
            f = os.path.join(directory, filename)
            try:
                crystalstructure, expanded, crystal, group = getlists(f)
                IDLIST.append(crystalstructure)
                ExpandedUnitCellBondDistances.append(expanded)
                CrystalCellBondDistances.append(crystal)
                SpaceGroup.append(group)
                count += 1
                pbar.update(1)
            except:
                print((filename) + " : FAILED")
```

```python
#ID SUB DOCUMENTS IN JSON
ChemicalName = []
Cell_Dimensions = []
Scattering_Components = []
Total_components_in_unit_cell = []
for i in IDLIST:
    string = str(i).split('\n')
    ChemicalName.append(string[0])
    Cell_Dimensions.append(string[1])
    Scattering_Components.append(string[2])
    Total_components_in_unit_cell.append(string[len(string)-1])
```

```python
#SPACE GROUP SUB DOCUMENTS IN JSON
SchoenfliesSymbolList = []
HermannMauginSymbolList = []
HallSymbolList = []
SgNumber = []
NumberOfSeitzMatrix = []
for i in SpaceGroup:
    string = str(i).split('\n')
    SchoenfliesSymbolList.append(string[1])
    HermannMauginSymbolList.append(string[2])
    HallSymbolList.append(string[3])
    SgNumber.append(string[4])
    NumberOfSeitzMatrix.append(string[5])
```

```python
import json

data = []

for i in range(len(IDLIST)):
    example = {
        "Crystal Information": {
            "ChemicalName": ChemicalName[i],
            "Cell_Dimensions": Cell_Dimensions[i],
            "Scattering_Components": Scattering_Components[i],
            "Total_components_in_unit_cell": Total_components_in_unit_cell[i]
        },
        "Space_Group_Information": {
            "Schoenflies_Symbol": SchoenfliesSymbolList[i].split("= ")[1],
            "Hermann-Maugin_Symbol": HermannMauginSymbolList[i].split("= ")[1],
            "Hall_Symbol": HallSymbolList[i].split("= ")[1],
            "Sg_Number": SgNumber[i].split("= ")[1],
            "Number_Of_Seitz_Matrix": NumberOfSeitzMatrix[i].split("= ")[1]
        },
        "Bond Distances": {
           # "Expanded Unit Cell": ExpandedUnitCellBondDistances[i].tolist(),
            "Unit Cell": CrystalCellBondDistances[i].tolist()
   
        }
    }
    data.append(example)

json_data = json.dumps(data, indent=4)

with open("/Users/seanwu/Desktop/data.json", "w") as json_file:
    json_file.write(json_data)
```
